```python
import pandas as pd
data = pd.read_csv(r"D:\Medium\Online_class_survey_.csv")
data
```

```python
data.drop('Unnamed: 0', axis = 1, inplace = True)
```

```python
data.dropna(inplace=True)
data.Faculty.value_counts()
```

```python
data.shape
```

```python
df = data.copy()
```

```python
print("Faculty: ", df['Faculty'].unique())
print("Bachelor Academic Year: ", df['Bachelor Academic Year'].unique())
print("Online classes: ", df['Online classes'].unique())
```

```python
from sklearn import preprocessing

le = preprocessing.LabelEncoder()

df['Faculty'] = le.fit_transform(df['Faculty'])
df[40:53]
```

```python
df['Bachelor Academic Year'] = le.fit_transform(df['Bachelor Academic Year'])
df[40:53]
```

```python
df['Faculty'] = data['Faculty']
df[40:53]
```

```python
ohe = preprocessing.OneHotEncoder()

X = ohe.fit_transform(df['Faculty'].values.reshape(-1, 1)).toarray()

faculty= df['Faculty'].unique()

df_ohe = pd.DataFrame(X, columns = ["Faculty_"+faculty[i] for i in range(len(faculty))])

df = pd.concat([df, df_ohe], axis = 1)
df.drop('Faculty', axis = 1, inplace = True)
df[40:53]
```

```python
Y = pd.get_dummies(df['Online classes'], prefix = 'Online classes')
df = pd.concat([df, Y], axis = 1)
df.drop('Online classes', axis = 1, inplace = True)
df[40:53]
```
