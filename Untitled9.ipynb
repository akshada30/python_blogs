```python
import folium
```

```python
import pandas as pd
```

```python
df = pd.read_csv("world.csv")
```

```python
df.drop("Unnamed: 0", axis=1, inplace=True)
```

```python
df
```

```python
avg = df.Cases.mean()
```

```python
map = folium.Map(zoom_start=4)
```

```python
df['Cases'].astype(int)
# df['Cases']=df['Cases']/df['Cases'].mean()
for lat, lon, value, name in zip(df['latitude'], df['longitude'], df['Cases'], df['Country']):
    folium.CircleMarker([lat, lon],
                        radius=value/avg,
                        popup = ('<strong>Country</strong>: ' + str(name).capitalize() + '<br>'
                                '<strong>Confirmed Cases</strong>: ' + str(value) + '<br>'),
                        color='red',
                        fill=True,
                        fill_color='red',
                        fill_opacity=0.3 ).add_to(map)
map
```
